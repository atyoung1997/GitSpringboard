#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None, parse_dates = True)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
#Set the properties index so that it recognizes the new column names, then transpose 
#and reset index so that we have numbered row indices instead of just the borough names

properties = properties.set_index(properties.columns[0]).transpose().reset_index()
properties.head()


```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#Rename the row index name, column index name, and then the first two column names

properties.index.name = 'Row Index'
properties.columns.name = 'Column Names'
properties.rename(columns = {properties.columns[0]: 'Borough Name', properties.columns[1]: 'Postal Code'}, inplace = True)
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Melt and then print the dataframe

properties = pd.melt(properties, id_vars = properties.columns[0:2], value_vars = properties.columns[2:], var_name = 'Month', value_name = 'Mean House Price')

properties.head()
properties.info()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Change borough name and postal code to all uppercase strings for consistency
# Change mean house price to all floats
# Month is already stored as datetime64

properties['Borough Name'] = properties['Borough Name'].astype(str, errors = 'raise').str.upper()
properties['Postal Code'] = properties['Postal Code'].astype(str, errors = 'raise').str.upper()
properties['Mean House Price'] = pd.to_numeric(properties['Mean House Price'], errors = 'raise')

properties.info()
properties
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Investigate the NaN Mean House price entries
# 930 NaN results, accounts for the difference between 14880 total results in the other columns and the 13950 non-null results in the mean house price column
# It appears that these null rows are labeled as UNNAMED for the borough name, and the postal code was a 'nan' string which was formatted by my transformations in the prior cell, which is why they did not show up as null in the .info() call

properties.loc[np.isnan(properties['Mean House Price']), :].info()
properties.loc[np.isnan(properties['Mean House Price']), :]

# drop all rows with NaN for the mean house price column, then print new summaries

properties.dropna(axis = 0, inplace = True)
properties.info()
properties.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# subset london only

cityoflondon_subset = properties.loc[properties['Borough Name'] == 'CITY OF LONDON', :]

# check that it makes sense

cityoflondon_subset.info()

# plot the data

cityoflondon_subset.plot(x = 'Month', y = 'Mean House Price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
# create year column, no need to use a lambda function

properties['Year'] = properties['Month'].dt.year

# check that each borough has a 1995-01-01 entry and a 2020-10-01 entry

print(
len(properties['Borough Name'].unique()),
len(properties.loc[properties['Month'] == '1995-01-01','Borough Name']),
len(properties.loc[properties['Month'] == '2020-10-01','Borough Name'])
)
 
# create new data frame from a dictionary with four columns: starting value, ending value, total change and average change

dict1 = {}

# calculate total time difference and date factor
startingdate = datetime.strptime('1995-01-01', '%Y-%m-%d')
endingdate = datetime.strptime('2020-10-01', '%Y-%m-%d')
datefactor = 1/((endingdate - startingdate).days/365.25)

for borough in properties['Borough Name'].unique():
    
    # define the values of new columns based on each borough name being iterated through
    startingvalue = float(properties.loc[(properties['Borough Name'] == borough) & (properties['Month'] == '1995-01-01'), 'Mean House Price'])
    endingvalue = float(properties.loc[(properties['Borough Name'] == borough) & (properties['Month'] == '2020-10-01'), 'Mean House Price'])
    totalchange = endingvalue/startingvalue
    averagechange = totalchange ** datefactor
    
    # add new values to dict1
    dict1[borough] = [startingvalue, endingvalue, totalchange, averagechange]

# create dataframe from dictionary
properties2 = pd.DataFrame.from_dict(dict1, columns = ['Starting Value', 'Ending Value', 'Total 10-Year Change', 'Average Yearly Change'], orient = 'index').sort_values(by = 'Average Yearly Change', ascending = False)
properties2

# The top five boroughs are Hackney, Lambeth, City of London, Tower Hamlets, Inner London





```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
def create_price_ratio():
    """ Calculate price ratio for a given borough, starting year, and ending year"""
    
    # take necessary inputs
    borough = str(input('Please enter the name of the borough.')).upper()
    startingyear = int(input('Please enter the starting year in YYYY Format.'))
    endingyear = int(input('Please enter the ending year in YYYY Format'))
    
    # perform subsetting and calculations
    startingmean = properties.loc[(properties['Borough Name'] == borough) & (properties['Year'] == startingyear), 'Mean House Price'].mean()
    endingmean = properties.loc[(properties['Borough Name'] == borough) & (properties['Year'] == endingyear), 'Mean House Price'].mean()
    priceratio = round((endingmean/startingmean-1)*100, 0)
    
    #print the result
    print('The average house price in ' + borough.title() + ' has increased roughly' + str(int(priceratio)) + ' percent from ' + str(startingyear) + ' to ' + str(endingyear) + '.')


create_price_ratio()  
    
    
    
    
    
    
    
    
    
```
